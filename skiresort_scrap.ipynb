```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
```

```python
URL = "https://www.skiresort.info/ski-resorts/pacific-states-west-coast/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36",
}
page = requests.get(URL, headers= headers)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
# soup
```

```python
resort_container = soup.find(id="resortList")
resort_urls = resort_container.find_all('a', class_="h3")
```

```python
resort_urls
```

```python
resort = []
for resort_url in resort_urls:
    # Save each resort information in the list resort
    
    resort_href = resort_url['href']
    if resort_href.startswith("http"):
        full_url = resort_href
    else:
        full_url = "https://www.skiresort.info" + resort_href

    try:
        resort_page = requests.get(full_url, headers=headers)
        resort_soup = BeautifulSoup(resort_page.content, "html.parser")
        
        name = resort_soup.find('h1').text.strip()
        elevation = resort_soup.find('div', class_="detail-links").text.strip()
        ski_lifts = resort_soup.find('div', class_="lift-info-group").text.strip()
        ##snow = resort_soup.find('div', class_="info-item-con no link").text.strip()
        ski_slopes = resort_soup.find('strong', id="selSlopetot").text.strip()
        ticket_price = resort_soup.find('td', id="selTicketA").text.strip()
        test_rating = resort_soup.find('span', class_="test").text.strip()

        resort.append([name, elevation, ski_lifts, ski_slopes, ticket_price, test_rating])
        

    ##except:
        ##resort.append([pd.NA, pd.NA, pd.NA, pd.NA]) 
    except Exception as e:
        print(f"Error on {full_url}: {e}")
        resort.append([pd.NA, pd.NA, pd.NA, pd.NA,pd.NA, pd.NA])
    
```

```python
df = pd.DataFrame(resort, columns = ['Resort name', 'Elevation', 'Ski lifts', 'Slopes', 'Ticket Price', 'Test Rating'])
df
```

```python
df["Resort name"] = df["Resort name"].astype(str).str.replace(r'\s*Ski resort\s*', '', regex=True)
df["Elevation"] = df["Elevation"].astype(str).str.extract(r'(\d+)')
df["Ski lifts"] = df["Ski lifts"].astype(str).str.extract(r'(\d+)')
df["Slopes"] = df["Slopes"].astype(str).str.extract(r'(\d+)')
df["Ticket Price"] = df["Ticket Price"].astype(str).str.extract(r'(\d+)')
df["Test Rating"] = df["Test Rating"].astype(str).str.extract(r'(\d+\.\d+)').astype(float)
df
```

```python
df.to_csv('Ski_Resort_Data2.csv', index=False)
```
